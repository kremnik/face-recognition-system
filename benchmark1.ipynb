```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
x = np.random.randn(2000, 8)
x = x - x.mean(axis=0)
x = x / x.std(axis=0)
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6)) = plt.subplots(nrows=2, ncols=3, figsize=(20,10), sharey="row")

ax1.scatter(x[:, 0], x[:, 1], s=3)

# y = kx + b
k = np.random.randn()
b = np.random.randn()
#x_ = np.linspace(x[:, 0].min(), x[:, 0].max(), len(x))
x_ = x[:, 0]
y_ = k * x_ + b

ax1.set_ylim(x[:, 1].min(), x[:, 1].max())
ax1.set_xlim(x[:, 0].min(), x[:, 0].max())
ax1.plot(x_, y_, c="g")

data_left = x[np.less(x[:, 1], y_)]
ax2.scatter(data_left[:, 0], data_left[:, 1], s=2, c="brown")

ax2.set_ylim(x[:, 1].min(), x[:, 1].max())
ax2.set_xlim(x[:, 0].min(), x[:, 0].max())
ax2.plot(x_, y_, c="g")

data_right = x[np.greater_equal(x[:, 1], y_)]
ax3.scatter(data_right[:, 0], data_right[:, 1], s=2, c="red")

ax3.set_ylim(x[:, 1].min(), x[:, 1].max())
ax3.set_xlim(x[:, 0].min(), x[:, 0].max())
ax3.plot(x_, y_, c="g")

data_right = data_right - data_right.mean(axis=0)
ax4.scatter(data_right[:, 0], data_right[:, 1], s=2, c="red")

data_right = data_right / data_right.std(axis=0)
ax5.scatter(data_right[:, 0], data_right[:, 1], s=2, c="red")

pca = PCA(n_components=2, whiten=True)
pca.fit(data_right)
data_right = pca.transform(data_right)
ax6.scatter(data_right[:, 0], data_right[:, 1], s=2, c="red")
```
